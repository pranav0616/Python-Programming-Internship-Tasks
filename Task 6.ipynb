```python
from tkinter import *
import random
from datetime import datetime
from tkinter import messagebox, filedialog

def total_bill():
    # Prices of items
    lassi_price = 50
    coffee_price = 20
    tea_price = 10
    juice_price = 30
    shakes_price = 50
    milk_price = 20
    shikanji_price = 15
    redbull_price = 150

    roti_price = 5
    dal_makhni_price = 120
    mutter_paneer_price = 150
    paratha_price = 40
    mix_veg_price = 70
    omelette_price = 20
    veg_biryani_price = 120
    rice_price = 50

    lassi_q = int(lassi_qty.get() or 0)
    coffee_q = int(coffee_qty.get() or 0)
    tea_q = int(tea_qty.get() or 0)
    juice_q = int(juice_qty.get() or 0)
    shakes_q = int(shakes_qty.get() or 0)
    milk_q = int(milk_qty.get() or 0)
    shikanji_q = int(shikanji_qty.get() or 0)
    redbull_q = int(redbull_qty.get() or 0)

    roti_q = int(roti_qty.get() or 0)
    dal_makhni_q = int(dal_makhni_qty.get() or 0)
    mutter_paneer_q = int(mutter_paneer_qty.get() or 0)
    paratha_q = int(paratha_qty.get() or 0)
    mix_veg_q = int(mix_veg_qty.get() or 0)
    omelette_q = int(omelette_qty.get() or 0)
    veg_biryani_q = int(veg_biryani_qty.get() or 0)
    rice_q = int(rice_qty.get() or 0)

    total_drinks_cost = (lassi_q * lassi_price + coffee_q * coffee_price +
                         tea_q * tea_price + juice_q * juice_price +
                         shakes_q * shakes_price + milk_q * milk_price +
                         shikanji_q * shikanji_price + redbull_q * redbull_price)

    total_foods_cost = (roti_q * roti_price + dal_makhni_q * dal_makhni_price +
                        mutter_paneer_q * mutter_paneer_price + paratha_q * paratha_price +
                        mix_veg_q * mix_veg_price + omelette_q * omelette_price +
                        veg_biryani_q * veg_biryani_price + rice_q * rice_price)

    service_charge = 10
    tax = (total_drinks_cost + total_foods_cost) * 0.08
    sub_total = total_drinks_cost + total_foods_cost + service_charge
    total_cost = sub_total + tax

    drinks_cost.set(total_drinks_cost)
    foods_cost.set(total_foods_cost)
    service_charge_cost.set(service_charge)
    paid_tax_cost.set(round(tax, 2))
    sub_total_cost.set(round(sub_total, 2))
    total_cost_var.set(round(total_cost, 2))

    date = datetime.now().date()
    bill_no = random.randint(100, 1000)
    bill_details.delete(1.0, END)
    bill_details.insert(END, f"Bill No: {bill_no}\t{date}\n")
    bill_details.insert(END, "===========================\n")
    bill_details.insert(END, "Items\t\tQty\tAmount\n")
    bill_details.insert(END, "===========================\n")
    if lassi_q > 0: bill_details.insert(END, f"Lassi\t\t{lassi_q}\t{lassi_q * lassi_price}\n")
    if coffee_q > 0: bill_details.insert(END, f"Coffee\t\t{coffee_q}\t{coffee_q * coffee_price}\n")
    if tea_q > 0: bill_details.insert(END, f"Tea\t\t{tea_q}\t{tea_q * tea_price}\n")
    if juice_q > 0: bill_details.insert(END, f"Juice\t\t{juice_q}\t{juice_q * juice_price}\n")
    if shakes_q > 0: bill_details.insert(END, f"Shakes\t\t{shakes_q}\t{shakes_q * shakes_price}\n")
    if milk_q > 0: bill_details.insert(END, f"Milk\t\t{milk_q}\t{milk_q * milk_price}\n")
    if shikanji_q > 0: bill_details.insert(END, f"Shikanji\t\t{shikanji_q}\t{shikanji_q * shikanji_price}\n")
    if redbull_q > 0: bill_details.insert(END, f"Redbull\t\t{redbull_q}\t{redbull_q * redbull_price}\n")
    if roti_q > 0: bill_details.insert(END, f"Roti\t\t{roti_q}\t{roti_q * roti_price}\n")
    if dal_makhni_q > 0: bill_details.insert(END, f"Dal Makhni\t\t{dal_makhni_q}\t{dal_makhni_q * dal_makhni_price}\n")
    if mutter_paneer_q > 0: bill_details.insert(END, f"Mutter Paneer\t\t{mutter_paneer_q}\t{mutter_paneer_q * mutter_paneer_price}\n")
    if paratha_q > 0: bill_details.insert(END, f"Paratha\t\t{paratha_q}\t{paratha_q * paratha_price}\n")
    if mix_veg_q > 0: bill_details.insert(END, f"Mix Veg\t\t{mix_veg_q}\t{mix_veg_q * mix_veg_price}\n")
    if omelette_q > 0: bill_details.insert(END, f"Omelette\t\t{omelette_q}\t{omelette_q * omelette_price}\n")
    if veg_biryani_q > 0: bill_details.insert(END, f"Veg Biryani\t\t{veg_biryani_q}\t{veg_biryani_q * veg_biryani_price}\n")
    if rice_q > 0: bill_details.insert(END, f"Rice\t\t{rice_q}\t{rice_q * rice_price}\n")
    bill_details.insert(END, "===========================\n")
    bill_details.insert(END, f"Total Cost:\t\t\t{total_cost}\n")

def reset_fields():
    lassi_qty.set("")
    coffee_qty.set("")
    tea_qty.set("")
    juice_qty.set("")
    shakes_qty.set("")
    milk_qty.set("")
    shikanji_qty.set("")
    redbull_qty.set("")
    roti_qty.set("")
    dal_makhni_qty.set("")
    mutter_paneer_qty.set("")
    paratha_qty.set("")
    mix_veg_qty.set("")
    omelette_qty.set("")
    veg_biryani_qty.set("")
    rice_qty.set("")

    drinks_cost.set("")
    foods_cost.set("")
    service_charge_cost.set("")
    paid_tax_cost.set("")
    sub_total_cost.set("")
    total_cost_var.set("")
    bill_details.delete(1.0, END)

root = Tk()
root.title("Restaurant Management System")
root.geometry("800x600")

# Variables
lassi_qty = StringVar()
coffee_qty = StringVar()
tea_qty = StringVar()
juice_qty = StringVar()
shakes_qty = StringVar()
milk_qty = StringVar()
shikanji_qty = StringVar()
redbull_qty = StringVar()
roti_qty = StringVar()
dal_makhni_qty = StringVar()
mutter_paneer_qty = StringVar()
paratha_qty = StringVar()
mix_veg_qty = StringVar()
omelette_qty = StringVar()
veg_biryani_qty = StringVar()
rice_qty = StringVar()

drinks_cost = StringVar()
foods_cost = StringVar()
service_charge_cost = StringVar()
paid_tax_cost = StringVar()
sub_total_cost = StringVar()
total_cost_var = StringVar()


drinks_frame = LabelFrame(root, text="Drinks", font=('arial', 12, 'bold'), fg="black", bd=10)
drinks_frame.pack(side=LEFT, fill="both", expand=True)

Checkbutton(drinks_frame, text="Lassi", variable=lassi_qty).grid(row=0, column=0, sticky=W)
Entry(drinks_frame, textvariable=lassi_qty).grid(row=0, column=1)
Checkbutton(drinks_frame, text="Coffee", variable=coffee_qty).grid(row=1, column=0, sticky=W)
Entry(drinks_frame, textvariable=coffee_qty).grid(row=1, column=1)
Checkbutton(drinks_frame, text="Tea", variable=tea_qty).grid(row=2, column=0, sticky=W)
Entry(drinks_frame, textvariable=tea_qty).grid(row=2, column=1)
Checkbutton(drinks_frame, text="Juice", variable=juice_qty).grid(row=3, column=0, sticky=W)
Entry(drinks_frame, textvariable=juice_qty).grid(row=3, column=1)
Checkbutton(drinks_frame, text="Shakes", variable=shakes_qty).grid(row=4, column=0, sticky=W)
Entry(drinks_frame, textvariable=shakes_qty).grid(row=4, column=1)
Checkbutton(drinks_frame, text="Milk", variable=milk_qty).grid(row=5, column=0, sticky=W)
Entry(drinks_frame, textvariable=milk_qty).grid(row=5, column=1)
Checkbutton(drinks_frame, text="Shikanji", variable=shikanji_qty).grid(row=6, column=0, sticky=W)
Entry(drinks_frame, textvariable=shikanji_qty).grid(row=6, column=1)
Checkbutton(drinks_frame, text="Redbull", variable=redbull_qty).grid(row=7, column=0, sticky=W)
Entry(drinks_frame, textvariable=redbull_qty).grid(row=7, column=1)

foods_frame = LabelFrame(root, text="Foods", font=('arial', 12, 'bold'), fg="black", bd=10)
foods_frame.pack(side=RIGHT, fill="both", expand=True)

Checkbutton(foods_frame, text="Roti", variable=roti_qty).grid(row=0, column=0, sticky=W)
Entry(foods_frame, textvariable=roti_qty).grid(row=0, column=1)
Checkbutton(foods_frame, text="Dal Makhni", variable=dal_makhni_qty).grid(row=1, column=0, sticky=W)
Entry(foods_frame, textvariable=dal_makhni_qty).grid(row=1, column=1)
Checkbutton(foods_frame, text="Mutter Paneer", variable=mutter_paneer_qty).grid(row=2, column=0, sticky=W)
Entry(foods_frame, textvariable=mutter_paneer_qty).grid(row=2, column=1)
Checkbutton(foods_frame, text="Paratha", variable=paratha_qty).grid(row=3, column=0, sticky=W)
Entry(foods_frame, textvariable=paratha_qty).grid(row=3, column=1)
Checkbutton(foods_frame, text="Mix Veg", variable=mix_veg_qty).grid(row=4, column=0, sticky=W)
Entry(foods_frame, textvariable=mix_veg_qty).grid(row=4, column=1)
Checkbutton(foods_frame, text="Omelette", variable=omelette_qty).grid(row=5, column=0, sticky=W)
Entry(foods_frame, textvariable=omelette_qty).grid(row=5, column=1)
Checkbutton(foods_frame, text="Veg Biryani", variable=veg_biryani_qty).grid(row=6, column=0, sticky=W)
Entry(foods_frame, textvariable=veg_biryani_qty).grid(row=6, column=1)
Checkbutton(foods_frame, text="Rice", variable=rice_qty).grid(row=7, column=0, sticky=W)
Entry(foods_frame, textvariable=rice_qty).grid(row=7, column=1)

bill_area_frame = Frame(root)
bill_area_frame.pack(side=TOP, fill="both", expand=True)
Label(bill_area_frame, text="Bill Details", font=('arial', 12, 'bold'), fg="black").pack()
bill_details = Text(bill_area_frame, height=15, width=50, bd=4)
bill_details.pack()


button_frame = Frame(root)
button_frame.pack(side=BOTTOM, fill="x")

Button(button_frame, text="Total", command=total_bill, font=('arial', 12, 'bold'), fg="black", bg="lightblue").grid(row=0, column=0, padx=5, pady=5)
Button(button_frame, text="Reset", command=reset_fields, font=('arial', 12, 'bold'), fg="black", bg="lightblue").grid(row=0, column=1, padx=5, pady=5)
Button(button_frame, text="Exit", command=root.quit, font=('arial', 12, 'bold'), fg="black", bg="lightblue").grid(row=0, column=2, padx=5, pady=5)


root.mainloop()
```
